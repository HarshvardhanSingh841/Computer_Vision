```python
import pandas as pd
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader, random_split
import matplotlib.pyplot as plt

# Custom Dataset Class for Image Data
class FashionDataset(Dataset):
    def __init__(self, file_path, transform=None):
        self.dataset = pd.read_csv(file_path)
        self.transform = transform

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, index):
        row_data = self.dataset.iloc[index]
        label = row_data.iloc[0]
        image_pixels = row_data.iloc[1:].values.astype(np.float32).reshape(28, 28)

        if self.transform:
            image_pixels = self.transform(image_pixels)

        image_tensor = torch.tensor(image_pixels)
        label_tensor = torch.tensor(label, dtype=torch.long)

        return image_tensor, label_tensor


# Load the CSV file and prepare the dataset
data_file = "train.csv"
full_dataset = FashionDataset(data_file)

# Split the dataset into training and validation sets
train_len = int(0.8 * len(full_dataset))
val_len = len(full_dataset) - train_len
train_data, val_data = random_split(full_dataset, [train_len, val_len])

# DataLoader for batch processing
train_loader = DataLoader(train_data, batch_size=64, shuffle=True)
val_loader = DataLoader(val_data, batch_size=64, shuffle=False)


# Function to visualize random images from the dataset
def display_samples(dataset, num_images):
    grid_dim = int(num_images ** 0.5)

    fig, axes = plt.subplots(grid_dim, grid_dim, figsize=(10, 10))
    axes = axes.flatten()

    for idx in range(num_images):
        img, lbl = dataset[idx]
        img = img.numpy()

        axes[idx].imshow(img, cmap='gray')
        axes[idx].set_title(f"Label: {lbl}")
        axes[idx].axis('off')

    plt.tight_layout()
    plt.show()


# Visualize 16 images from the training dataset
display_samples(train_data, num_images=16)


# Mapping of numerical labels to clothing categories
class_labels = {
    0: "T-Shirt/Top",
    1: "Trouser",
    2: "Pullover",
    3: "Dress",
    4: "Coat",
    5: "Sandal",
    6: "Shirt",
    7: "Sneaker",
    8: "Bag",
    9: "Ankle Boot",
}

# Organize images by their labels
images_by_label = {label: [] for label in range(10)}

for i in range(len(full_dataset)):
    img, lbl = full_dataset[i]
    lbl = lbl.item()  # Convert tensor to int for indexing
    if len(images_by_label[lbl]) < 10:  # Limit to 10 images per label
        images_by_label[lbl].append(img.numpy())

# Plot grouped images for each label
fig, axes = plt.subplots(10, 6, figsize=(16, 12))
plt.subplots_adjust(wspace=0.6, hspace=0.6)

for label in range(10):
    # Add label names in the first column
    axes[label, 0].text(
        0.5, 0.5, class_labels[label], fontsize=10, ha='center', va='center', transform=axes[label, 0].transAxes
    )
    axes[label, 0].axis('off')

    # Display 10 images corresponding to each label
    for j in range(5):
        axes[label, j + 1].imshow(images_by_label[label][j], cmap='gray')
        axes[label, j + 1].axis('off')

plt.show()
```

```python
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader, random_split

class Rawset(Dataset):
    def __init__(self, csv_file, transform=None):
        """
        Dataset for raw pixel data and labels.
        """
        self.data = pd.read_csv(csv_file)
        self.transform = transform

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]
        label = row.iloc[0]  # First column is the label
        features = row[1:].values.astype(np.float32) / 255.0  # Normalize pixel values

        if self.transform:
            features = self.transform(features)

        return torch.tensor(features), torch.tensor(label, dtype=torch.long)

# Paths to CSV files
train_csv = "train.csv"
test_csv = "test.csv"

# Load datasets
raw_train_dataset = Rawset(train_csv)
raw_test_dataset = Rawset(test_csv)

# Split training dataset into train and validation
train_size = int(0.8 * len(raw_train_dataset))
val_size = len(raw_train_dataset) - train_size
train_dataset, val_dataset = random_split(raw_train_dataset, [train_size, val_size])

# Create DataLoaders
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False)
test_loader = DataLoader(raw_test_dataset, batch_size=64, shuffle=False)

```

```python
class RawNet(nn.Module):
    def __init__(self, input_dim, num_classes):
        super(RawNet, self).__init__()
        self.network = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(64, num_classes)
        )

    def forward(self, x):
        return self.network(x)
```

```python

import wandb
import torch.optim as optim
from torch.nn import CrossEntropyLoss

wandb.login()
# Initialize WandB
wandb.init(project="RawNet-Train-eval", config={
    "epochs": 20,
    "batch_size": 64,
    "learning_rate": 0.001
})
config = wandb.config

# Model Setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
input_dim = train_dataset[0][0].shape[0]
num_classes = len(pd.read_csv("train.csv").iloc[:, -1].unique())

model = RawNet(input_dim, num_classes).to(device)
criterion = CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=config.learning_rate)

# Training Function
def train_one_epoch(epoch):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()

    wandb.log({"Train Loss": running_loss / len(train_loader),
               "Train Accuracy": 100. * correct / total,
               "Epoch": epoch})

# Validation Function
def validate(epoch):
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0

    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)

            running_loss += loss.item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

    wandb.log({"Validation Loss": running_loss / len(val_loader),
               "Validation Accuracy": 100. * correct / total,
               "Epoch": epoch})

# Training Loop
for epoch in range(config.epochs):
    train_one_epoch(epoch)
    validate(epoch)
from sklearn.metrics import classification_report, accuracy_score


def evaluate_model(model, dataloader, device):
    """
    Evaluate the model on a given dataloader and calculate metrics.
    
    Args:
    - model (torch.nn.Module): The trained model.
    - dataloader (torch.utils.data.DataLoader): DataLoader for evaluation.
    - device (torch.device): Device to run the model on (CPU or CUDA).
    
    Returns:
    - metrics (dict): Dictionary containing evaluation metrics.
    """
    model.eval()  # Set model to evaluation mode
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for images, labels in dataloader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, preds = torch.max(outputs, 1)  # Get the predicted classes

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Calculate metrics
    accuracy = accuracy_score(all_labels, all_preds)
    report = classification_report(all_labels, all_preds, output_dict=True, zero_division=0)

    # Extract F1 score, precision, recall for each class
    class_metrics = {
        f"{key}_precision": value["precision"]
        for key, value in report.items()
        if key not in ["accuracy", "macro avg", "weighted avg"]
    }
    class_metrics.update({
        f"{key}_recall": value["recall"]
        for key, value in report.items()
        if key not in ["accuracy", "macro avg", "weighted avg"]
    })
    class_metrics.update({
        f"{key}_f1": value["f1-score"]
        for key, value in report.items()
        if key not in ["accuracy", "macro avg", "weighted avg"]
    })

    # Log metrics to WandB
    wandb.log({"Test Accuracy": accuracy})
    for metric, value in class_metrics.items():
        wandb.log({f"Test/{metric}": value})

    # Print a summary report
    print("Classification Report:")
    print(classification_report(all_labels, all_preds, zero_division=0))

    # Return metrics for further use
    return {"accuracy": accuracy, **report}
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Evaluate the model
metrics = evaluate_model(model, test_loader, device)

# Print overall accuracy
print(f"Test Accuracy: {metrics['accuracy']:.4f}")

wandb.finish()

```

```python
class CannyNet(nn.Module):
    def __init__(self, input_dim, num_classes):
        super(CannyNet, self).__init__()
        self.network = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(64, num_classes)
        )

    def forward(self, x):
        return self.network(x)
```

The main reason i think that RAWnet performed so well is because i think the image resolution is too low , leading to lower pixels and input dimensions and thereby allowing discrete images to be detected by just feeding the flattened images I dont think such a strategy will work for the more complicated datasets which might require further pre processing for correct use
Similar to this i think the reason canny edge failed is because of the low resolution images not giving reliable edges to feed for classification , shirt and coat have similar edge maps in such low res images making them hard to distinguish,

Scenarios where Rawnet might underperform:
its working on raw pixel data thereby it might be harder for it to learn relevant patterns in more complexly oriented images (inverted etc) also causing it to struggle with factors like lighting and scale variations.

For cannynet
the features provided should prove to be more semantically meaningful as edges help in capturing more structural information however may miss texture based data captured in rawnet

HOG seems to be the best of both worlds as it can capture texture data and also gradient orientation(local edge) allowing MLP to learn more generalisable features as compared to rawnet and outperform cannynet when faced with classification requiring fine grained texture recognition.
